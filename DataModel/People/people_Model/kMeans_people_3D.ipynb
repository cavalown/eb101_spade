```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
# Though the following import is not directly being used, it is required
# for 3D projection to work
from mpl_toolkits.mplot3d import Axes3D

from sklearn.cluster import KMeans
from sklearn import datasets
```

```python
np.random.seed(500)

get_ipython().run_line_magic('matplotlib', 'inline')
```

```python
# Load data
data_all = pd.read_csv('peopleSample_combind.csv')
```

```python
data=data_all[['Axis1','Axis2','Axis3','Axis4']]
```

```python
# 設定 模型 估計參數
estimators = [('k_means_3', KMeans(n_clusters=3)),
              ('k_means_4', KMeans(n_clusters=4)),
              ('k_means_5', KMeans(n_clusters=5)),
              ('k_means_6', KMeans(n_clusters=6))]
```

```python
# 資料建模 並 視覺化 結果
fignum = 1
titles = ['3 clusters', '4 clusters', '5 clusters','6 clusters']

for name, est in estimators:
    fig = plt.figure(fignum, figsize=(12, 9))
    ax = Axes3D(fig, rect=[0, 0, .95, 1], elev=48, azim=134)
    
    ## fit data
    est.fit(data)
    
    labels = est.labels_

    ax.scatter(data['Axis1'],data['Axis2'],data['Axis3'],
               c=labels.astype(np.float), edgecolor='k')

    ax.w_xaxis.set_ticklabels([])
    ax.w_yaxis.set_ticklabels([])
    ax.w_zaxis.set_ticklabels([])
    ax.set_xlabel('Feature1')
    ax.set_ylabel('Feature2')
    ax.set_zlabel('Feature3')
    ax.set_title(titles[fignum - 1])
    ax.dist = 10
    fignum = fignum + 1
```
