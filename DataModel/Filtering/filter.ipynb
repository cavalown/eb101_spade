```python
import pandas as pd
import numpy as np
import heapq
```

```python
dataP = pd.read_csv('people_clusterNu_df.csv')
dataD = pd.read_csv('clusterNu_df.csv')
```

```python
dataP = dataP[["Calories","Moisture","Protein","Saturated_fat","Carbohydrate","Dietary_fiber","Mg","Fe","Zn","VitaminA","VitaminB_group","Fatty_acid_S","Fatty_acid_M","Fatty_acid_P"]]
dataP
```

```python
dataD = dataD[["Calories","Moisture","Protein","Saturated_fat","Carbohydrate","Dietary_fiber","Mg","Fe","Zn","VitaminA","VitaminB_group","Fatty_acid_S","Fatty_acid_M","Fatty_acid_P"]]
dataD
```

```python
dataPP = np.array(dataP)
dataPP[0]
```

```python
dataDD = np.array(dataD)
dataDD[1]
```

```python
# 使用scipy套件 計算各種距離
from scipy.spatial.distance import pdist
```

#### Eucledian Distance

```python
peopleCluster=['ClusterA','ClusterB','ClusterC','ClusterD']
pc=0
for eachPeople in dataPP:
    allED = dict()
    dc=1
    for eachDish in dataDD:
        X=np.vstack([eachPeople,eachDish])
        ED=pdist(X)
        distance = ED.item(0)
        allED[distance]= 'Dish cluster'+str(dc)
        dc+=1
        
    minDD = heapq.nsmallest(7,allED)
    print(peopleCluster[pc],':')
    for minD in minDD:
        print(allED[minD])
    pc+=1
    print('-'*50)
```

#### Manhattan Distance

```python
peopleCluster=['ClusterA','ClusterB','ClusterC','ClusterD']
pc=0
for eachPeople in dataPP:
    allED = dict()
    dc=1
    for eachDish in dataDD:
        X=np.vstack([eachPeople,eachDish])
        ED=pdist(X,'cityblock')
        distance = ED.item(0)
        allED[distance]= 'Dish cluster'+str(dc)
        dc+=1
        
    minDD = heapq.nsmallest(7,allED)
    print(peopleCluster[pc],':')
    for minD in minDD:
        print(allED[minD])
    pc+=1
    print('-'*50)
```

### Cosine Distance

```python
peopleCluster=['ClusterA','ClusterB','ClusterC','ClusterD']
pc=0
for eachPeople in dataPP:
    allED = dict()
    dc=1
    for eachDish in dataDD:
        X=np.vstack([eachPeople,eachDish])
        ED=pdist(X,'cosine')
        distance = ED.item(0)
        allED[distance]= 'Dish cluster'+str(dc)
        dc+=1
        
    minDD = heapq.nsmallest(7,allED)
    print(peopleCluster[pc],':')
    for minD in minDD:
        print(allED[minD])
    pc+=1
    print('-'*50)
```
